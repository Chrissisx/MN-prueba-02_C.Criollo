## Determinante

```python
import copy

%autoreload 2
from src import (
    eliminacion_gaussiana,
    matriz_aumentada,
    separar_m_aumentada,
    descomposicion_LU,
    
)
def calc_determinante(A: list[list[float]]) -> float:
    """Función que calcula el determinante usando el método...
    [eliminación gaussiana, Gauss-Jordan, Gauss-Jacobi o Gauss-Seidel]

    ## Parameters
    ``A``: Matriz cuadrada de tamaño n x n

    ## Return
    ``detA``: Determinante de la matriz A

    """
    
    # Eliminación Gaussiana para calcular el determinante

    n = len(A)
    M = copy.deepcopy(A)
    det = 1
    for i in range(n):
        # Buscar el máximo en la columna i para evitar división por cero
        max_row = max(range(i, n), key=lambda r: abs(M[r][i]))
        if M[max_row][i] == 0:
            return 0
        if max_row != i:
            M[i], M[max_row] = M[max_row], M[i]
            det *= -1  # Cambio de signo por intercambio de filas
        det *= M[i][i]
        for j in range(i + 1, n):
            factor = M[j][i] / M[i][i]
            for k in range(i, n):
                M[j][k] -= factor * M[i][k]
    return det
```
